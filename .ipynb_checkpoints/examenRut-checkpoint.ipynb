   6. (5 p) **Distribucin binomial**

   Un proceso tiene probabilidad de xito *p* = 0.2 en cada ensayo. Si realizas *n* = 10 ensayos,

   a. Cul es la probabilidad de obtener exactamente 3 xitos?


   Para la distribucin binomial donde:
   - \( n \) es el nmero de ensayos,
   - \( p \) es la probabilidad de xito en cada ensayo,
   - \( k \) es el nmero exacto de xitos.

   ##### **La frmula para la probabilidad de obtener exactamente \( k \) xitos:**
 
   $$
   P(X = k) = \binom{n}{k} \cdot p^k \cdot (1 - p)^{n - k}
   $$

   Donde:

   $$
   \binom{n}{k} = \frac{n!}{k!(n - k)!}
   $$

   Para el inciso a. tenemos:
   $$
   \binom{10}{3} = \frac{10!}{3!(10 - 3)!}= 120
   $$
   Por lo que: 

   $$
   P(X = 3) = \binom{10}{3} \cdot 0.2^3 \cdot (1 - 0.2)^{10 - 3}= 120*0.008*0.2097= 0.2013 
   $$

   R/ La probabilidad de obtener exactamente 3 xitos con una probabilidad de 0.2 y 10 ensayos es de  0.2013 o 20.13%.
 
   b. Cul es la probabilidad de obtener al menos 1 xito?
   Para calcular la probabilidad de tener al menos un xito tenemos que calcular la probabilidad de tener 0 xitos y se la restamos a 1.

   $$
   P(X \geq 1) = 1 - P(X = 0)
   $$

   $$
   \binom{10}{0} = \frac{10!}{0!(10 - 0)!}=1
   $$

   $$
   P(X = 0) = \binom{10}{0} \cdot 0.2^0 \cdot (1 - 0.2)^{10 - 0}= 1*1*0.8^{10}\approx 0.1074
   $$


   R/ La probabilidad de que ocurra al menos un xito es de 1-0.1074 $\approx$ 0.8926 o 89.26%.

8. (8 p) **Intervalo de confianza para la media*

   Una muestra de tamao *n* = 25 arroja una media muestral de 80 y desviacin estndar muestral de 10.

   a. Calcula el intervalo de confianza al 95 % para la media poblacional (asume distribucin normal).

   b. Interpreta el resultado.


   #### a. Clculo del intervalo de confianza al 95 %

   - Tamao de la muestra: \( n = 25 \)
   - Media muestral: \( $\bar{x}$ = 80 \)
   - Desviacin estndar muestral: \( s = 10 \)

   Como \( n < 30 \), asumimos que la variable a analizar es cuantitativa y la distribucin es "paramtrica" usamos la **distribucin t de Student** con \( df = n - 1 = 24 \).  
   Un nivel de confianza del 95 %, implica un 5% de error total (=0.05). Como el intervalo es simtrico, ese error se reparte en ambos extremos de la distribucin: /2=0.025 a cada lado. As se obtiene el valor crtico _{/2} para calcular el intervalo. Apoyndonos de una tabla, el valor crtico \( t_{0.025, 24} $\approx$ 2.064 \).

   #### Frmula del intervalo de confianza:

   $$
   IC = \bar{x} \pm t \cdot \frac{s}{\sqrt{n}}
   $$

   #### Sustituyendo los valores:

   $$
   IC = 80 \pm 2.064 \cdot \frac{10}{\sqrt{25}} = 80 \pm 2.064 \cdot 2 = 80 \pm 4.128
   $$

   $$
   IC_{95\%} = [75.872,\ 84.128]
   $$

   Podemos decir que, **con un 95 % de confianza**, la media real de la poblacin (por ejemplo, un parmetro como la puntuacin promedio de calidad de vida o nivel de anticuerpos en mujeres cubanas con lupus) se encuentra **entre 75.87 y 84.13**.

14. (4 p) **Definicin de OR**
    a. Define el odds ratio.
 
   El odds ratio (OR) es una medida estadstica que se usa para comparar las probabilidades (odds) de un evento entre dos grupos.

   Se calcula a partir de una tabla de contingencia 22:
   $$
   \text{OR} = \frac{a \cdot d}{b \cdot c}
   $$

   Donde:

   |                        | Evento S | Evento No | Total |
   |------------------------|-----------|-----------|--------|
   | Grupo Expuesto         |     a     |     b     | a + b  |
   | Grupo No Expuesto      |     c     |     d     | c + d  |

  
   b. Explica qu significa un OR = 1, OR > 1 y OR < 1.

   OR = 1: No hay asociacin entre la exposicin y el evento. Las probabilidades son iguales en ambos grupos.

   OR > 1:Hay mayor probabilidad del evento en el grupo expuesto. Ejemplo: si OR = 2, el evento es el doble de probable en ese grupo.

   OR < 1:Hay menor probabilidad del evento en el grupo expuesto. Ejemplo: si OR = 0.5, el evento es la mitad de probable en ese grupo.
